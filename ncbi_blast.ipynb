## Threads

```python
import multiprocessing as mp
from time import sleep
```

```python
len_max = 10000
len_min = 2000
```

```python
def Nblast(i, Seq_id, TxID):
    sleep(i*20)
    result_handle =NCBIWWW.qblast("blastn", "nr", Seq_id, entrez_query='NOT txid'+TxID+'[ORGN]')
    with open(Seq_id+ "_" +TxID +".xml", "w") as save_to:
        save_to.write(result_handle.read())
        result_handle.close()
        print(i)
```

```python
def multicore(Pool=230):
  pool = mp.Pool(processes=Pool)
  Num = 0
  for i in Seq_dic:
        for ii in Insect_dict:
            Num += 1
            # Working function "echo" and the arg 'i'
            multi_res = [pool.apply_async(Nblast,(Num,i,ii,))]
  pool.close()
  pool.join()

if __name__ == '__main__':
  multicore()
```

```python
## Read all gi ID
GI_ID = []
for i in List:
    result=open(i,"r")
    records= NCBIXML.parse(result)
    item=next(records)
    for alignment in item.alignments:
        for hsp in alignment.hsps:
            if hsp.expect < 0.01:
                #print('****Alignment****')
                #print('sequence:', alignment.title)
                #print('length:', alignment.length)
                #print('score:', hsp.score)
                #print(hsp.query[0:90] + '...')
                #print(hsp.match[0:90] + '...')
                #print(hsp.sbjct[0:90] + '...')
                #print(len(hsp.sbjct))
                if alignment.length <= len_max and alignment.length >= len_min:
                    print(alignment.title)
                    print(alignment.length)
                    GI_ID += [alignment.title]
```

```python
GI_ID_uq = list(set(GI_ID))
# Find all spcise
R_list = ["PREDICTED: ", "Synthetic construct "]
Sp_all = [i.split("| ")[-1].replace(R_list[0], "").replace(R_list[1], "") for i in GI_ID_uq]
# Clear for some unexpected format: 
rep_dic = {i:i[i.find(' ')+1:] for i in Sp_all if "|" in i}
[Sp_all.remove(i) for i in rep_dic]
[Sp_all.append(rep_dic[i])  for i in rep_dic]
SP_list = list(set([ " ".join(i.split(" ")[:2]) for i in Sp_all]))
```

```python
def is_chinese(string):
    for ch in string:
        if u'\u4e00' <= ch <= u'\u9fff':
            return True
    return False
```
